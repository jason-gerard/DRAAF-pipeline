```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
x = [] # BER
ys = {} # Corrected BERs

with open("./ber_table.csv", "r") as f:
    lines = f.readlines()
    for ber in lines[0].split(",")[1:]:
        x.append(float(ber))

    for line in lines[1:]:
        parts = line.split(",")
        ys[parts[0]] = [float(ber) for ber in parts[1:]]

x.reverse()
for y in ys:
    ys[y].reverse()
print(x)
print(ys)
```

```python
output_dir = "output_graphs"

label = "Corrected BER"

fig = plt.figure()
ax = fig.add_subplot(111)

for codec in ys:
    y = ys[codec]
    plt.plot(x, y, label=codec, clip_on=False)

plt.ylabel(label)
plt.xlabel("BER")
plt.legend(loc="lower left")
plt.grid(linestyle='-', color='0.95')

plt.yscale("log")
plt.xscale("log")

plt.ylim(0.000000001, 1)
plt.xlim(min(x), max(x))

plt.savefig(os.path.join(output_dir, f"{label.replace(' ', '_').replace('/', '_')}.png"))
plt.show()
```
